```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np

#load the file
df=pd.read_csv('climate_action_data.csv')
```

```python
df.head(20)
```

```python

df.info()
```

```python

#Data Types of each column
print(df.dtypes)
```

```python
#Number of rows and columns
print(f'Rows & Columns: \n{df.shape}')
```

```python
#Checking for Null Values
df.isnull().sum()
```

```python
#Handling Null Values
df['Crop_Type']=df['Crop_Type'].fillna(df['Crop_Type'].mode()[0])
```

```python
#Changing data type of date column
df['Date']=pd.to_datetime(df['Date'])
print(df.info())
```

```python
#Change to numeric values
#Handle missing values
numeric_columns =['Soil_Moisture(%)','Soil_pH','Temperature(C)','Humidity(%)','Irrigation_Recommended(mm)','Fertilizer_Recommended(kg/ha)']
for col in numeric_columns:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce') 
        df[col] = df[col].fillna(df[col].median())
        
print("Missing values after handling:")
print(df.isnull().sum())
```

```python
print(df.to_string())
```

```python
print(df.isnull().sum())
```

```python
#Null dates 
df['Date']=pd.to_datetime(df['Date'],errors='coerce')
print(df.isnull().sum())
```

```python
#Replace missing values with the next non-null value below it in the column using 'bfill:
df['Date'].fillna(method='bfill', inplace=True)
```

```python

print(df.isnull().sum())
```

```python
df.duplicated()
```

```python
df=df.drop_duplicates()
print(f'Number of rows after dropping duplicate rows: {df.shape[0]}')
```

```python

df.duplicated()
```

```python
#Clean Dataset
print(df.isnull().sum())
print(df.duplicated().sum())
```

```python

#Correlation Heat Map
corr_matrix=df[['Soil_Moisture(%)','Soil_pH','Temperature(C)','Humidity(%)','Fertilizer_Recommended(kg/ha)','Irrigation_Recommended(mm)']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix,cmap='RdBu',annot=True, fmt='.2f', linewidths=0.5, center=0)
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.groupby('Crop_Type')['Fertilizer_Recommended(kg/ha)'].mean().sort_values(ascending=False)
```

```python
#histograms for Soil_Moisture and Soil_pH Distribution
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.title("Soil_Moisture(%) Distribution")
sns.histplot(df['Soil_Moisture(%)'],bins=30,kde=True)

plt.subplot(1,2,2)
plt.title("Soil_pH Distribution")
sns.histplot(df['Soil_pH'],bins=30,kde=True)
plt.show()
```

```python
#Histograms for Temperature and Humidity Distribution
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.title("Temperature(C) Distribution")
sns.histplot(df['Temperature(C)'],bins=30,kde=True)

plt.subplot(1,2,2)
plt.title("Humidity(%) Distribution")
sns.histplot(df['Humidity(%)'],bins=30,kde=True)
plt.show()
```

```python
#Histograms for Fertilizer and Irrigation
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.title("Fertilizer_Recommended(kg/ha) Distribution")
sns.histplot(df['Fertilizer_Recommended (kg/ha)'],bins=30,kde=True)

plt.subplot(1,2,2)
plt.title("Irrigation_Recommended(mm) Distribution")
sns.histplot(df['Irrigation_Recommended(mm)'],bins=30,kde=True)
plt.show()
```

```python
#From the correlation heatmap, we can see that:
#Soil_Moisture(%) has a weak positive correlation with Soil_pH (0.08) and Temperature(C) (0.07).
#Soil_pH has a weak positive correlation with Temperature(C) (0.05) and Humidity(%) (0.04).
#Temperature(C) has a weak positive correlation with Humidity(%) (0.06).
#this shows that these environmental variables vary independently
```

```python
#The crop type with the highest average soil moisture:

max_soil_moisture=df.groupby('Crop_Type')['Soil_Moisture(%)'].mean()
crop=max_soil_moisture.idxmax()
max_value=max_soil_moisture.max()
print(f'crop  with max average value: \n{crop} ={max_value}')
```

```python
#Appropriate irrigation adjustments for crops experiencing average temperatures above 30C:
average_temp=df.groupby('Crop_Type')['Temperature(C)'].mean()
crops=average_temp
print(crops)
```

```python
#Crops above temperature of 30C
above_30=average_temp[average_temp>30]
print(above_30)
```

```python
#Appropriate irrigation adjustments for crops experiencing average temperatures above 30C:
#Continuous irrigation is recommended for crops experiencing average temperatures above 30C to prevent water stress and ensure optimal growth conditions.
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.title("Average temperature vs crops")
sns.barplot(y=average_temp, x=above_30)
plt.show()
```
